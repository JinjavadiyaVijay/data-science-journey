```python
class product:
    count = 0  
    def __init__(self, name, price):
        self.name = name
        self.price = price
        product.count += 1  

    def display(self):
        print(f"Product Name: {self.name}, Price: ${self.price:.2f}")

@staticmethod
def discount(price, persentage):
    return price - (price * persentage / 100)


item = product("Laptop", 999.99)
item.display()
item2 = product("Smartphone", 499.49)
item2.display()
print(f"Total products created: {product.count}")

print(discount(item.price, 10))
```

```python
class bankAccount :
    accoucnt_number = 0
    owner_name = ""
    balance = 0.0

    def deposit(self,amount):
        self.balance += amount
        print(f"Deposited: ${amount:.2f}. New balance: ${self.balance:.2f}")

    def withdraw(self,amount):
        self.balance -= amount
        print(f"Withdrew: ${amount:.2f}. New balance: ${self.balance:.2f}")

    def chack_balance(self):
        print(f"Current balance: ${self.balance:.2f}")


my_account = bankAccount()
my_account.deposit(1500)
my_account.withdraw(300)
my_account.chack_balance()
```

```python
class book:
    title = ""
    author = ""
    list_of_review = []


    def add_new_review(self, review):
        self.list_of_review.append(review)
        print(f"Added review: {review}")
        
    def count_review(self):
        print(f"Total reviews: {len(self.list_of_review)}")   

    def display_reviews(self):
        print("Reviews:")
        for review in self.list_of_review:
            print(f"- {review}")

my_book = book()
book.title = "The Great Gatsby"
book.author = "F. Scott Fitzgerald"
my_book.add_new_review("A fascinating exploration of the American Dream.")
my_book.add_new_review("Beautifully written with complex characters.")

my_book.count_review()

my_book.display_reviews()
```

```python
class student:
    __name = ""
    __roll_number = 0
    __marks = []

    def set_name(self,name):
        if isinstance(name, str):
            self.__name = name
        else:
            print("Invalid name")

    def get_name(self):
        return self.__name
    
    def set_roll_number(self,roll_number):
        if roll_number <100 and roll_number >0:
            self.__roll_number = roll_number
        else:
            print("Invalid roll number")

    def get_roll_number(self):
        return self.__roll_number
    
    def add_mark(self,mark):
        if mark >=0 and mark <=100:
            self.__marks.append(mark)
    def get_marks(self):
        return self.__marks

s1 = student()
s1.set_name("Alice")
s1.set_roll_number(97)
s1.add_mark(85)
s1.add_mark(90)
print(f"Student Name: {s1.get_name()}")
print(f"Roll Number: {s1.get_roll_number()}")
print(f"Marks: {s1.get_marks()}")
    
```

```python
import math
class shape:
    def area(self):
        pass

class rectangle(shape):
    
    def __init__(self,lenth,width):
        self.length = lenth
        self.width = width
    def area(self):
        return self.length * self.width

class circle(shape):
    def __init__(self,r):
        self.redus = r
        def area(self,):
            return math.pi * self.redus * self.redus

class triangle(shape):
    def __init__(self,base,height):
        self.base = base
        self.height = height
    def area(self):
        return 0.5 * self.base * self.height        
    
r = rectangle(5,10)
print(f"Rectangle Area: {r.area()}")

c = circle(7)
print(f"Circle Area: {c.area()}")

t = triangle(6,8)
print(f"Triangle Area: {t.area()}")
```

```python
from abc import ABC, abstractmethod

class Employee(ABC):
    @abstractmethod
    def salary(self):
        pass

class Intern(Employee):
    def salary(self):
        return 15000

class FullTimeEmployee(Employee):
    def salary(self):
        return 35000

class ContractEmployee(Employee):
    def salary(self):
        return 32000

intern = Intern()    
print(f"Intern Salary: ${intern.salary()}")

full_time = FullTimeEmployee()
print(f"Full-Time Employee Salary: ${full_time.salary()}")

contract = ContractEmployee()
print(f"Contract Employee Salary: ${contract.salary()}")
```
