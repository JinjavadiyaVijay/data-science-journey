```python
import pandas as pd
import numpy as np
```

```python
class linearRegression :
    def __init__(self,learning_rate = 0.1,n_itr = 1000):
        self.bais = None # intercept
        self.weight = None # feature
        self.lr = learning_rate
        self.itr = n_itr

    def fit(self,x,y):

        m, n = x.shape
        
        self.bais = 0
        self.weight = np.zeros(n)
        
        for i in range(self.itr):

            y_pred = self.bais +np.dot(x,self.weight)
        
            db = (1/m)*np.sum(y_pred-y)
            dw = (1/m) * np.dot(x.T, (y_pred - y))
        
            self.bais = self.bais - self.lr*db
            self.weight = self.weight - self.lr*dw
        
    def predict(self,x):
        y_pred = y_pred = self.bais +np.dot(x,self.weight)
        return y_pred
    
```

```python
x = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

model = linearRegression()

model.fit(x,y)

output = model.predict(x)

print(output)
```
