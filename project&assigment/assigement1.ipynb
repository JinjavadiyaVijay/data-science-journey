```python
import pandas as pd
import numpy as np
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
URL = "https://quotes.toscrape.com/page/1/"
count=1

while True:
    res = requests.get(f"https://quotes.toscrape.com/page/{count}/")
    soup = BeautifulSoup(res.text,"lxml")
    qoute = soup.select("div.quote")

    if not qoute:
        print("not valid")    
        break
    else:
        print(f"dowloaded page{count}")
    with open(f"scaped_data/quote{count}.html","w") as f:
        f.write(res.text)
    count=count+1

```

```python
c = 1
info1 = []
for i in range(10):
    with open(f"scaped_data/quote{c}.html", "r") as f:
        soup = BeautifulSoup(f.read(), "lxml")

    quotes = soup.select("div.quote")
    for quote in quotes:
        tags = [tag.get_text(strip=True) for tag in quote.select("a.tag")]
        for t in tags:
            if t == "life":
                q = quote.select_one("span.text").get_text(strip=True)
                a = quote.select_one("small.author").get_text(strip=True)
                print(f"Quote: {q}, Author: {a}, Tags: {tags},page :{c}")
                info1.append([q,a])
            else:
                continue
    c += 1
```
