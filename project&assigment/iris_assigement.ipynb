```python
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,precision_score,recall_score,r2_score
from sklearn.linear_model import LogisticRegression
```

```python
iris_data = pd.read_csv("Iris.csv")
iris_data.head()
```

```python
print(iris_data["Species"].unique())
```

```python
x = iris_data.drop("Species",axis=1)
y = iris_data["Species"]

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_encoded = le.fit_transform(y)

x_train,x_test,y_train,y_test = train_test_split(
    x,y_encoded,random_state=42
)

```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
logistic_reg = LogisticRegression(max_iter=100)

from sklearn.pipeline import Pipeline

pipe1 = Pipeline([

("sacaler",scaler),
("logistic_reg",logistic_reg)

])

pipe1.fit(x_train,y_train)
y_pred1 =pipe1.predict(x_test)

```

```python
from sklearn.metrics import (
    confusion_matrix, accuracy_score,
    precision_score, recall_score, f1_score
)

cm = confusion_matrix(y_test, y_pred1)
print("Confusion Matrix:\n", cm)

accuracy  = accuracy_score(y_test, y_pred1) * 100
precision = precision_score(y_test, y_pred1, average='macro') * 100
recall    = recall_score(y_test, y_pred1, average='macro') * 100
f1        = f1_score(y_test, y_pred1, average='macro') * 100

print(f"Accuracy  : {accuracy:.2f}%")
print(f"Precision : {precision:.2f}%")
print(f"Recall    : {recall:.2f}%")
print(f"F1 Score  : {f1:.2f}%")
```

```python
from sklearn.naive_bayes import GaussianNB 
GaussianNB_model = GaussianNB()

pipe2 = Pipeline([

("sacaler",scaler),
("gaussianNB",GaussianNB_model)

])

pipe2.fit(x_train,y_train)
y_pred2 =pipe2.predict(x_test)
```

```python
from sklearn.metrics import (
    confusion_matrix, accuracy_score,
    precision_score, recall_score, f1_score
)

cm = confusion_matrix(y_test, y_pred2)
print("Confusion Matrix:\n", cm)

accuracy  =  accuracy_score(y_test, y_pred2) * 100
precision = precision_score(y_test, y_pred2, average='macro') * 100
recall    =    recall_score(y_test, y_pred2, average='macro') * 100
f1        =        f1_score(y_test, y_pred2, average='macro') * 100

print(f"Accuracy  : {accuracy:.2f}%") 
print(f"Precision : {precision:.2f}%")
print(f"Recall    : {recall:.2f}%")
print(f"F1 Score  : {f1:.2f}%")
```
