```python
import os
import shutil
import random

# your folders
cats_src = r"C:\Users\jinju\Documents\AI_ML\Prime\Deep_Learing\prac_manual\data\assignment\cats_set"
dogs_src = r"C:\Users\jinju\Documents\AI_ML\Prime\Deep_Learing\prac_manual\data\assignment\dogs_set"

# new dataset path
base_path = r"C:\Users\jinju\Documents\AI_ML\Prime\Deep_Learing\prac_manual\data\dataset"

train_path = os.path.join(base_path, "train")
test_path  = os.path.join(base_path, "test")

# create folders
for path in [train_path, test_path]:
    os.makedirs(os.path.join(path, "cats"), exist_ok=True)
    os.makedirs(os.path.join(path, "dogs"), exist_ok=True)

# split function
def split_data(src, train_dst, test_dst, split=0.8):
    files = os.listdir(src)
    random.shuffle(files)

    split_idx = int(len(files) * split)

    train_files = files[:split_idx]
    test_files  = files[split_idx:]

    for f in train_files:
        shutil.copy(os.path.join(src, f), os.path.join(train_dst, f))

    for f in test_files:
        shutil.copy(os.path.join(src, f), os.path.join(test_dst, f))

# apply split
split_data(cats_src, os.path.join(train_path, "cats"), os.path.join(test_path, "cats"))
split_data(dogs_src, os.path.join(train_path, "dogs"), os.path.join(test_path, "dogs"))

print("Dataset ready ✅")
```

```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))
])

train_data = datasets.ImageFolder(
    root=r"C:\Users\jinju\Documents\AI_ML\Prime\Deep_Learing\prac_manual\data\dataset\train",
    transform=transform
)

test_data = datasets.ImageFolder(
    root=r"C:\Users\jinju\Documents\AI_ML\Prime\Deep_Learing\prac_manual\data\dataset\test",
    transform=transform
)

trainloader = DataLoader(train_data, batch_size=32, shuffle=True)
testloader = DataLoader(test_data, batch_size=32, shuffle=False)
```

```python
import torch 
import torch.nn as nn
class CNN(nn.Module):
    def __init__(self):
        super (CNN,self).__init__()

        self.convo_lyr = nn.Sequential(

            nn.Conv2d(3,32,kernel_size = 3,padding = 1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),

            nn.Conv2d(32,64,kernel_size=3,padding = 1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),

            nn.Conv2d(64,128,kernel_size = 3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),

           
        )
    
        self.fully_conn_lyr = nn.Sequential(
            nn.Linear(100352, 256),
            nn.ReLU(),

            nn.Linear(256,2)
        )
        
    def forward(self, x):
        x = self.convo_lyr(x)
        print(x.shape)   # 👈 check this once
        x = x.view(x.size(0), -1)
        output = self.fully_conn_lyr(x)
        return output
```

```python
import torch.optim as optim
model = CNN()
critarion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())
```

```python
epochs = 10 

for  epoch in range(epochs):
    epoch_loss = 0.0

    for Image,labels in trainloader:
        optimizer.zero_grad()

        output = model(Image)
        loss = critarion(output,labels)
        loss.backword()
        optimizer.step()

        epoch_loss += loss.item() 

        print("epoch : ",epoch,"||","epoch_train_lost :",epoch_loss/len(trainloader))
```

```python
model.eval()
correct=0.0
total=0.0
test_loss = 0.0
with torch.no_grad():
       for Image,labels in testloader:
        output_test = model.forward(Image)
        loss = critarion(output_test,labels)
        test_loss += loss.item()
             
        __,predict = torch.max(output_test,1)
        correct += (predict == labels).sum().item()
        total += labels.size(0) 
```
