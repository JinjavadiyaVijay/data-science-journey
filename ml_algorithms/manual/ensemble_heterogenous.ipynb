```python
from sklearn.datasets import make_classification,make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,r2_score,precision_score,classification_report

```

```python

X, y = make_regression(
    n_samples = 1500,
    n_features = 20,
    n_informative = 5,
    random_state = 42
)
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    X,y,test_size=0.3,random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lin = LinearRegression()
dt =DecisionTreeRegressor(max_depth=3)
svm = SVR()
```

```python
from sklearn.ensemble import VotingRegressor

v_reg_ = VotingRegressor(
    estimators=[
        ("l1", lin),
        ("dt", dt),
        ("svm ", svm)
    ]
)
```

```python
v_reg_.fit(x_train,y_train)

v_reg_.fit(x_train,y_train)

y_train_pred = v_reg_.predict(x_train)
y_test_pred = v_reg_.predict(x_test)

print("r2_score : ",r2_score(y_train,y_train_pred))
print("r2_score : ",r2_score(y_test,y_test_pred))
```

# Stacking

```python
X, y = make_classification(
    n_samples = 1500,
    n_features = 20,
    n_informative = 5,
    random_state = 42
)
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    X,y,test_size=0.3,random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
lg = LogisticRegression()
dt = DecisionTreeClassifier(max_depth=3)
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
from sklearn.ensemble import StackingClassifier

stk = StackingClassifier(
    estimators=[
        ("lg",lg),
        ("dt",dt),
        ("knn",knn)
    ],
    cv=5
)
```

```python
stk.fit(x_train,y_train)

y_pred_trian = stk.predict(x_train)
y_test_pred = stk.predict(x_test)

print("classification_report train : ",classification_report(y_train,y_pred_trian))
print("classification_report : ", classification_report(y_test,y_test_pred))
```
