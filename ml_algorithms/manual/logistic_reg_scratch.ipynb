```python
import pandas as pd 
import numpy as np
```

```python
class LogisticRegression :
    def __init__(self,n_itr = 1000, lr = 0.01):
        self.bais=None
        self.weight = None
        self.lr = lr
        self.itr = n_itr

    def sigmoid(self,z):
        return 1/(1+np.exp(-z))
     
    def fit(self,X,y):

        m, n = X.shape

        self.bais = 0
        self.weight = np.zeros(n) 

        for i in range(self.itr):

            z = self.bais + np.dot(X,self.weight)
            y_pred = self.sigmoid(z)


            db = 1/m*np.sum(y_pred-y)
            dw = 1/m*np.dot(X.T,(y_pred-y))

            self.bais = self.bais-db*self.lr
            self.weight = self.weight-dw*self.lr

    def probability (self,X) :
        z = self.bais + np.dot(X, self.weight)
        return self.sigmoid(z)

    def predict(self,X,threshold = 0.5) :
        prediction = self.probability(X)
        y_pred_bool = prediction >= threshold
        return y_pred_bool.astype(int)
    
```

```python
X = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])
y = np.array([0, 0, 1, 1])

model = LogisticRegression()

model.fit(X,y)
print(model.probability(X))
print(model.predict(X))
```
