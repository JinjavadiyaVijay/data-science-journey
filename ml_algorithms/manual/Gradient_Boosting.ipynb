```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import r2_score,precision_score,accuracy_score
from sklearn import datasets
```

```python
X,y= datasets.make_regression(
    n_features=13,
    n_samples=1000,
    noise=50,
    random_state=42
)
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    X,y,random_state=42,test_size=0.2
)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
GB_reg = GradientBoostingRegressor(n_estimators=300,learning_rate=0.05,max_depth=3
                                   ,subsample=0.80)

GB_reg.fit(x_train,y_train)
y_train_pred =GB_reg.predict(x_train)
y_test_pred = GB_reg.predict(x_test)

print("train r2_score : ",r2_score(y_train,y_train_pred))
print("test r2_score : ",r2_score(y_test,y_test_pred))
```

# classification

```python
X,y = datasets.make_classification(n_samples=700,n_features=20,n_informative = 10,random_state=42)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42,)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GB_claaci_ = GradientBoostingClassifier(max_depth=3,n_estimators=200,learning_rate=0.09)

GB_claaci_.fit(x_train,y_train)

y_train_pred = GB_claaci_.predict(x_train)
y_test_pred = GB_claaci_.predict(x_test)

print("train r2_score : ",r2_score(y_train,y_train_pred))
print("test r2_score : ",r2_score(y_test,y_test_pred))
```
