```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score
```

```python
df = pd.read_csv("1-heart.csv")
```

```python
df.head()
df.info()
df.isnull().sum()
df.isna
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns="target"),df["target"],random_state=42,test_size=0.2)
```

```python
y_train[y_train == 1]
y_train[y_train == 0] #make sure data has eqal split 
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000) #by default max iteraction is 100

model.fit(x_train,y_train)
y_pred = model.predict(x_test)

print(f"accuracy score:{accuracy_score(y_test,y_pred)}")
print(f"precision score:{precision_score(y_test,y_pred)}")
```
