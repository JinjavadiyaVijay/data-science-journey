#OneHotEncoding

```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("insurance.csv")
df.shape
```

```python
df.columns
```

```python
x = pd.get_dummies(df, columns=["region"], drop_first=True, dtype=int)
x["sex"] = x["sex"].map({"female": 1, "male": 0})
x["smoker"] = x["smoker"].map({"yes": 1, "no": 0})
```

# Interaction

x["age_smoker"] = x["age"]*x["smoker"]
