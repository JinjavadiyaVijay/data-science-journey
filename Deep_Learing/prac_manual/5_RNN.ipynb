## Pre-processing

```python
# converting data to lower case

df['review'] =df["review"].str.lower()

```

```python
# Removing the URLs

import re  # re helps computers understand and manipulate text patterns automatically.

def remove_url (text):
    text = re.sub(r"http\S+","",text) #(pattern , repl, string) \S -> non-white space \s -> white space 
    return text
df['review'] = df['review'].apply(remove_url)
```

```python
# removing punctuations 
def remove_pun_ (text):
    text = re.sub(r"[A-Za-z0-9\s]","",text) # if value is not in A-Z a-z 0-9 \s its called punctuation
    return text

df['review'] = df['review'].apply(remove_pun_)
```

```python
#remove html tags 

def remove_tags(text):
    text = re.sub(r"<.*?>","",text) # here . mean any character * means infinate char size ? means non greedy manner 
    return text
df['review'] = df['review'].apply(remove_tags)
```
