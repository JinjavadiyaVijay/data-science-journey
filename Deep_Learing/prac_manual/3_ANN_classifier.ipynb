```python
X = df.drop("Class", axis=1)
y = df["Class"]
```

```python
df["Class"].unique()
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder

sclr = StandardScaler()
X = sclr.fit_transform(X)

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(
    X,y,random_state=74
)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset 
```

```python
x_train_tensor = torch.tensor(X_train,dtype = torch.float32)
y_train_tensor = torch.tensor(y_train,dtype = torch.long)

x_test_tensor = torch.tensor(X_test, dtype = torch.float32)
y_test_tensor = torch.tensor(y_test, dtype = torch.long)
```

```python
x_train_dataset = TensorDataset(x_train_tensor,y_train_tensor)
x_test_dataset = TensorDataset(x_test_tensor,y_train_tensor)

train_loader = DataLoader(x_train_dataset,shuffle=True,batch_size=32)
test_loader = DataLoader(x_test_dataset,shuffle=True,batch_size=32)
```

```python
X_train.shape[0]
```

```python
class ANN(nn.Module) :

    def __init__(self):
        super(ANN,self).__init__()

        self.model = nn.Sequential(
           
           #input layer
            nn.Linear(X_train.shape[1],64),
            nn.ReLU(),

            #1 hidden layer
            nn.Linear(64,64),
            nn.ReLU(),

            #2 hidden layer
            nn.Linear(64,64),
            nn.ReLU(),

            #final layer 
            nn.Linear(64,7)
        )

    def forward(self,x):
            return self.model(x)
        
```

```python
model = ANN()
criteria = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())


epoch = 200
total_loss = 0.0

for epoch in range(epoch):
    model.train()
    runing_loss = 0.0
    for xb,yb in train_loader:
        
        optimizer.zero_grad()
        output = model(xb)
        loss = criteria(output,yb)
        loss.backward()
        optimizer.step()

        runing_loss += loss.item()
    
    total_loss =runing_loss / len(train_loader)
     
    print(f"epoch = {epoch}/{epoch}, loss = {total_loss}")
```

```python

# Evaluate
model.eval()

total = 0
correct = 0

with torch.no_grad():
    for xb, yb in test_loader:
        outputs = model(xb) # [0.2, 0.5, 1.3, -0.5, ..] - 7 vals
        
        _, predicted = torch.max(outputs, 1) # Returns TWO things: 1.Maximum value     2.Index of maximum value
        correct += (predicted == yb).sum().item()
        total += yb.size(0) # actual samples     in each batch

print("accuracy: ", correct/total * 100)

```
